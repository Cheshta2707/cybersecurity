```python
df_cleaned['risk_score']=(
    (df_cleaned['failed_logins']>5).astype(int)
    +(df_cleaned['login_attempts']>8).astype(int)
    +(df_cleaned['unusual_time_access']==1).astype(int)
    +(df_cleaned['ip_reputation_score_least_trustworthy']<0.3).astype(int)
)
```

```python
def risk(score):
    if score >= 3:
        return 'High'
    elif score >= 2:
        return 'Medium'
    return 'Low'
df_cleaned['risk_level'] = df_cleaned['risk_score'].apply(risk)
```

```python
sns.countplot(
    data=df_cleaned,
    x='risk_level'
)
plt.show()
```
